```python
# Se sustituye el valor númerico que dan título a las columnas por su título correspondiente:

df_1.columns =(['Variance', 'Skewness', 'Kurtosis', 'Entropy', 'Class'])
df_1
```

```python
#Se seleccionan las variables independientes y las dependientes:

x = df_1.iloc[:,0:4].values
y = df_1.iloc[:,4].values
```

```python
# Se divide el dataframe en la matriz de entrenamiento y la matriz de test:


from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2, random_state=65)
```

```python
# Número de registros de entrenamiento

print(len(x_train))
```

```python
# Número de registros de test

print(len(x_test))
```

```python
# Escalado de los datos del dataframe

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_train=sc.fit_transform(x_train)
x_test_sc=sc.fit_transform(x_test)
```

```python
print(len(x_train))
```

SE DEFINE LA RED NEURONAL

```python
from keras.optimizers import Adadelta

red = Sequential()

    #Agregado de capa de entrada
    
red.add(Dense( units = 6, activation = 'relu', kernel_initializer='uniform'))

    #Agregado de capa oculta
    
red.add(Dense( units = 2, activation = 'relu',  kernel_initializer='uniform'))
red.add(Dense( units = 8, activation = 'relu',  kernel_initializer='uniform'))
red.add(Dense( units = 8, activation = 'relu',  kernel_initializer='uniform'))

     #Agregado de capa de salida
    
red.add(Dense( units = 1, activation = 'sigmoid',  kernel_initializer='uniform'))

    #Compilado de la red
    

opt = 'Adam'
red.compile(optimizer=opt, loss="binary_crossentropy", metrics=['accuracy'])

```

```python
# Entrenamiento del modelo
    
red.fit(x_train, y_train, batch_size=10, epochs=(100))
```

```python
# Ejecución de la spredicciones

predict = red.predict(x_test)
predict
```

```python
# Para poder ejecutar la métrica y comprobar el porcentaje de acierto es necesario redondear los valores de las predicciones a la unidad para poder comparalos

predict_sc=np.round(predict)

predict_sc
```

```python
# Medición del porcentaje de acierto

from sklearn.metrics import confusion_matrix

nc = confusion_matrix(y_test,predict_sc)
nc
```
